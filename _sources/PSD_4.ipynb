```python
import os

# Tentukan direktori tempat file WAV berada
direktori = 'suara'

# List semua file dalam direktori
files = os.listdir(direktori)

# Dictionary untuk melacak berapa kali kata "angry" muncul dalam nama file
count_dict = {}

# Loop melalui setiap file dalam direktori
for file in files:
    if file.endswith('.wav') and 'ps' in file:  # Pastikan itu adalah file WAV dan mengandung kata "angry"
        old_name = os.path.join(direktori, file)
        
        # Hitung berapa kali kata "angry" muncul dalam nama file
        count = count_dict.get('ps', 0)
        
        # Membuat nama baru dengan angka yang bertambah
        new_name = os.path.join(direktori, f'ps{count + 1}.wav')
        count_dict['ps'] = count + 1
        
        os.rename(old_name, new_name)
        print(f'Mengganti nama {old_name} menjadi {new_name}')
```

```python
import librosa
import numpy as np
import os
import pandas as pd
from scipy.stats import kurtosis, skew

data_fitur = []


direktori_audio = 'suara' 


for filename in os.listdir(direktori_audio):
    if filename.endswith('.wav'):
        # Membaca file WAV
        file_path = os.path.join(direktori_audio, filename)
        y, sr = librosa.load(file_path, sr=None)

        # Menghitung mean, median, dan standar deviasi
        mean_value = np.mean(y)
        median_value = np.median(y)
        std_deviation = np.std(y)
        
        # Menghitung kurtosis dan skewness
        kurtosis_value = kurtosis(y)
        skewness_value = skew(y)

        # Mendapatkan label berdasarkan nama file atau sumber data lainnya
        # Misalnya, jika nama file mengandung label emosi, Anda dapat mengambilnya dari nama file.
        label_emosi = filename.split('_')[0]  # Contoh: "senang_audio.wav" akan menghasilkan label "senang"

        # Menyimpan hasil dalam dictionary
        fitur_audio = {
            'Nama File': filename,
            'Mean': mean_value,
            'Median': median_value,
            'Std Deviasi': std_deviation,
            'Kurtosis': kurtosis_value,
            'Skewness': skewness_value,

        }

        # Menambahkan dictionary ke list data_fitur
        data_fitur.append(fitur_audio)

# Membuat DataFrame Pandas dari data_fitur
df = pd.DataFrame(data_fitur)

# Menyimpan data ke file CSV jika diperlukan
df.to_csv('keluar.csv', index=False)
```

```python
# example of fitting a model on the scaled dataset
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from pickle import dump
# prepare dataset
X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=1)
# split data into train and test sets
X_train, _, y_train, _ = train_test_split(X, y, test_size=0.33, random_state=1)
# define scaler
scaler = MinMaxScaler()
# fit scaler on the training dataset
scaler.fit(X_train)
# transform the training dataset
X_train_scaled = scaler.transform(X_train)
# define model
k = 5  # Ganti dengan jumlah tetangga yang Anda inginkan
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train_scaled, y_train)
# save the model
dump(model, open('model_asli.pkl', 'wb'))
# save the scaler
dump(scaler, open('scaler.pkl', 'wb'))
```

```python
# load model and scaler and make predictions on new data
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from pickle import load
# prepare dataset
X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=1)
# split data into train and test sets
_, X_test, _, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# load the model
model = load(open('model_asli.pkl', 'rb'))
# load the scaler
scaler = load(open('scaler.pkl', 'rb'))
# check scale of the test set before scaling
print('Raw test set range')
for i in range(X_test.shape[1]):
 print('>%d, min=%.3f, max=%.3f' % (i, X_test[:, i].min(), X_test[:, i].max()))
# transform the test dataset
X_test_scaled = scaler.transform(X_test)
print('Scaled test set range')
for i in range(X_test_scaled.shape[1]):
 print('>%d, min=%.3f, max=%.3f' % (i, X_test_scaled[:, i].min(), X_test_scaled[:, i].max()))
# make predictions on the test set
yhat = model.predict(X_test_scaled)
# evaluate accuracy
acc = accuracy_score(y_test, yhat)
print('Test Accuracy:', acc)
```
